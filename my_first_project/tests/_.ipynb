```python
import duckdb

# Aponte diretamente para o arquivo que o dbt criou
# Baseado no seu sys.path anterior, o caminho é este:
caminho_banco = '/Users/baia/Desktop/PYTHON/repo-dsa/data-engineer/my_first_project/my_local_duckdb.db'

print(f"Conectando em: {caminho_banco}")

# connect(read_only=True) é uma boa prática para leitura para evitar travar o banco se o dbt estiver rodando
con = duckdb.connect(caminho_banco, read_only=True)

# Verificar tabelas
print("\nTabelas disponíveis:")
con.sql("SHOW TABLES").show()
```

```python
query = "SELECT * FROM analise_clientes_premium"

try:
    con.sql(query).show()
except Exception as e:
    print(f"Erro ao ler tabela: {e}")
```

```python
print("\nDados:")
try:
    con.sql("SELECT * FROM clients").show()
except Exception as e:
    print(f"Erro ao ler tabela: {e}")
```
